```python
import pandas as pd
import numpy as np
from re import findall
```

```python
ws = pd.read_csv('oportunidades.csv')
```

```python
andrea = pd.read_excel('db_PPF_validacao_para UNB_ FINAL.xlsx')
```

```python
comum=list(set(ws['opo_titulo']).intersection(andrea['opo_titulo']))
```

```python
comum = ws[ws['opo_titulo'].isin(comum)]
```

```python
ws['classificado'] = ['']*len(ws)
for i in ws['opo_titulo']:
    if i in list(comum['opo_titulo']) :
        ws.at[ws.loc[ws['opo_titulo'] == i].index, 'opo_brazil'] = comum.loc[comum['opo_titulo'] == i]['opo_brazil']
        ws.at[ws.loc[ws['opo_titulo'] == i].index, 'classificado'] = 'S'
```

```python
A = pd.read_excel('oportunidadesrotulo.xlsx')
B = A[['codigo', 'opo_dificuldade']]
ws = ws.merge(B, on='codigo', how='left')
```

```python
ws.to_csv('classificados.csv')
```

```python
faltantes = ws.loc[~ws['opo_dificuldade'].isin([1,2,3])]

keywords = [
    'brazil', 'latin america', 'underdeveloped', 'south america', 'brics',
    'mercosul', 'portuguese', 'middle income'
]
keywords = '(' + '|'.join(keywords) + ')'

for i in faltantes.index:
    texto = str(faltantes.loc[i]['opo_texto'] + faltantes.loc[i]['opo_texto_ele'])
    if(findall(keywords, texto.lower())):
        faltantes.at[i,'opo_brazil'] = 'Y'
    else:
        faltantes.at[i,'opo_brazil'] = 'N'
```

```python
faltantes.loc[faltantes['opo_brazil'] == 'N']
```

```python
faltantes = faltantes.loc[faltantes['opo_brazil'] == 'N']

keywords = [
    'self-funded'
]
keywords = '(' + '|'.join(keywords) + ')'

for i in faltantes.index:
    texto = str(faltantes.loc[i]['opo_texto'] + faltantes.loc[i]['opo_texto_ele'])
    if(findall(keywords, texto.lower())):
        faltantes = faltantes.drop(i)
```
